```python
# Loading Libraries

import requests
import bs4 as bs
import numpy as np
import pandas as pd
```

```python
source1 = requests.get("https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory").content
source1

df_covid = pd.read_html(source1,attrs={'class':'wikitable sortable tpl-blanktable'})[0]
df_covid
```

```python
# Cleaning

for column in ["Cases", "Deaths", "Deathsper million"]:
    df_covid[column] = df_covid[column].str.replace(r"[^0-9]", '', regex=True)
    df_covid[column] = df_covid[column].apply(pd.to_numeric)

df_covid = df_covid[:-1]
# df_covid["recovered"] = df_covid["Cases"] - df_covid["Deaths"] 
df_covid.drop(columns=["Country","Unnamed: 5", "Unnamed: 6", "Unnamed: 7"], inplace=True)
df_covid
```

```python
df_covid.rename(index=str, columns={"Country.1": "country", "Deathsper million": "deaths_per_million", "Deaths": "deaths"}, inplace=True)

df_covid['country'] = df_covid['country'].str.replace(r"\[\w*\]+", '', regex=True)

df_covid = df_covid.dropna() 
df_covid = df_covid.set_index('country')
df_covid['cases_per_deaths'] = df_covid['Cases']/df_covid['deaths']
df_covid = df_covid.sort_values(by = 'cases_per_deaths', ascending = False)
df_covid.head(20)
```

While the metrics cases, or deaths are a portion of the total population, this metric namely cases_per_deaths tries to give a better understanding of a country's healthcare systems, where one would want the deaths to be lower and recovery higher, which are both inferred from the healthcare systems of a country.
